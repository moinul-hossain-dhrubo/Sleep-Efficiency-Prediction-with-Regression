##About Dataset:
####The dataset contains information about a group of test subjects and their sleep patterns. Each test subject is identified by a unique "Subject ID" and their age and gender are also recorded. The "Bedtime" and "Wakeup time" features indicate when each subject goes to bed and wakes up each day, and the "Sleep duration" feature records the total amount of time each subject slept in hours. The "Sleep efficiency" feature is a measure of the proportion of time spent in bed that is actually spent asleep. The "REM sleep percentage", "Deep sleep percentage", and "Light sleep percentage" features indicate the amount of time each subject spent in each stage of sleep. The "Awakenings" feature records the number of times each subject wakes up during the night. Additionally, the dataset includes information about each subject's caffeine and alcohol consumption in the 24 hours prior to bedtime, their smoking status, and their exercise frequency.

The main objective of this project is to develop a predictive model that can estimate sleep efficiency based on various features. Sleep efficiency is a measure of the effectiveness and quality of sleep, and accurately predicting it can provide valuable insights into an individual's sleep patterns.

To achieve this goal, three different regression techniques were employed: Linear Regression, Ridge Regression, and Lasso Regression. These methods utilize statistical algorithms to establish relationships between the sleep efficiency variable and the corresponding features.

After conducting the analysis, it was observed that both Linear Regression and Ridge Regression (with a regularization parameter, or alpha, set to 0.08) exhibited similar performance in predicting sleep efficiency. Based on these findings, it can be concluded that both Linear Regression and Ridge Regression are viable options for predicting sleep efficiency in this project. However, further research and experimentation may be needed to determine the optimal model for different datasets or specific contexts.

Here are the equations for Linear Regression, Ridge Regression, and Lasso Regression, along with their corresponding loss functions:

1. Linear Regression:
Linear regression aims to find the best-fit line that minimizes the sum of squared errors between the predicted values and the actual values. The equation for linear regression can be written as:

y = β0 + β1x1 + β2x2 + ... + βnxn + ε

where:

y is the dependent variable (in this case, sleep efficiency)
x1, x2, ..., xn are the independent variables (features)
β0, β1, β2, ..., βn are the regression coefficients (to be estimated)
ε is the error term
The loss function for linear regression is the Mean Squared Error (MSE), which measures the average squared difference between the predicted and actual values:

Loss = (1/n) * Σ(yi - (β0 + β1x1 + β2x2 + ... + βnxn))^2

2. Ridge Regression:
Ridge regression is an extension of linear regression that adds a penalty term to the loss function to reduce overfitting. The equation for ridge regression is:

y = β0 + β1x1 + β2x2 + ... + βnxn + ε

The loss function for ridge regression is the MSE with an additional L2 regularization term:

Loss = (1/n) * Σ(yi - (β0 + β1x1 + β2x2 + ... + βnxn))^2 + λ * Σ(βi)^2

where:

λ is the regularization parameter (controls the strength of regularization)
Σ(βi)^2 sums up the squared values of the regression coefficients

3. Lasso Regression:
Lasso regression is another extension of linear regression that incorporates a penalty term to encourage sparsity in the model. The equation for lasso regression is:

y = β0 + β1x1 + β2x2 + ... + βnxn + ε

The loss function for lasso regression is the MSE with an additional L1 regularization term:

Loss = (1/n) * Σ(yi - (β0 + β1x1 + β2x2 + ... + βnxn))^2 + λ * Σ|βi|

where:

λ is the regularization parameter (controls the strength of regularization)
Σ|βi| sums up the absolute values of the regression coefficients

In ridge regression, the L2 regularization term encourages small but non-zero coefficients, while in lasso regression, the L1 regularization term encourages sparse solutions by promoting some coefficients to be exactly zero.



##Importing Libraries

```python
# Everyday data science libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Model libraries, scaling, metrics, etc
from sklearn.preprocessing import StandardScaler

# for building linear regression models and preparing data
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

##Data Preprocessing

```python
#Dropping ID column which is unnecessary
df = df.drop('ID',axis=1)
```

```python
#Converting categorical datas to numerical
df['Smoking status'] = df['Smoking status'].map({'Yes': 1, 'No': 0})

#Gender
df['Gender'] =df['Gender'].map({'Male':1,'Female':0})
```

```python
#Converting bedtime to Epoch

df['Bedtime'] = pd.to_datetime(df['Bedtime'], format='%Y-%m-%d %H:%M:%S')
df['Bedtime'] = df['Bedtime'].apply(lambda x: int(x.timestamp()))
df['Wakeup time'] = pd.to_datetime(df['Wakeup time'], format='%Y-%m-%d %H:%M:%S')
df['Wakeup time'] = df['Wakeup time'].apply(lambda x: int(x.timestamp()))
```

```python
#Finding Null

print(df.isnull().sum())
```

##Scaling Features

```python
# Initialize the class
scaler_linear = StandardScaler()

# Compute the mean and standard deviation of the training set then transform it
X = scaler_linear.fit_transform(X)
```

#Model Building

```python
#Training the model

from sklearn.model_selection import train_test_split



X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state=50)
```

```python
# Initialize the class
linear_model = LinearRegression()

# Train the model
linear_model.fit(X_train, y_train)
```

```python
# Feed the scaled training set and get the predictions
yhat = linear_model.predict(X_train)

# Use scikit-learn's utility function and divide by 2
train_mse = mean_squared_error(y_train, yhat) / 2
train_r2 = r2_score(y_train,yhat)

print(f"training MSE : {train_mse}")
print(f"training R2 :{train_r2}")
```

```python
yhat_test = linear_model.predict(X_test)

# Use scikit-learn's utility function and divide by 2
test_mse = mean_squared_error(y_test, yhat_test) / 2
test_r2 = r2_score(y_test,yhat_test)

print(f"Cross validation MSE: {test_mse}")
print(f"Cross validation R2 :{test_r2}")
```

##Model Comparison

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

models = [LinearRegression(),Ridge(),Lasso()]

def compare_models_train_test():
  for model in models:
    #training the model
    model.fit(X_train,y_train)
    #evaluating the model
    test_data_prediction = model.predict(X_test)
    accuracy = r2_score(y_test, test_data_prediction)
    mse = mean_squared_error(y_test,test_data_prediction) / 2
    
    print('R^2 score of the model', model, '=', accuracy)
    print('MSE score of the model', model, '=', mse)

compare_models_train_test()
```
